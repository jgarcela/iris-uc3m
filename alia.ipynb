```python
from datetime import datetime
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

# Modelo
model_id = "BSC-LT/salamandra-2b-instruct"
tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    device_map="auto",
    torch_dtype=torch.bfloat16
)


# Definir categorías
categories = ["Hombres", "Mujeres", "Mixto", "Neutro"]

# Construir mensaje con instrucciones mejoradas y formato de respuesta
message = [
    { "role": "system", "content": f"Clasifica el siguiente texto en una de estas categorías según el enfoque principal del contenido. "
                                  f"Las categorías son: 'Hombres' si la noticia habla principalmente de hombres, "
                                  f"'Mujeres' si habla principalmente de mujeres, 'Mixto' si menciona tanto a hombres como a mujeres, "
                                  f"y 'Neutro' si no habla de personas o está relacionado con temas generales. "
                                  f"Por favor, responde con un formato JSON con una sola clave: 'category', "
                                  f"y el valor debe ser el nombre de la categoría correspondiente." },
    { "role": "user", "content": text }
]

# Generar fecha
date_string = datetime.today().strftime('%Y-%m-%d')

# Crear el prompt con el template del modelo
prompt = tokenizer.apply_chat_template(
    message,
    tokenize=False,
    add_generation_prompt=True,
    date_string=date_string
)

# Tokenizar y generar salida
inputs = tokenizer.encode(prompt, add_special_tokens=False, return_tensors="pt")
outputs = model.generate(input_ids=inputs.to(model.device), max_new_tokens=50)  # Limitar la respuesta

# Decodificar y limpiar la salida
output_text = tokenizer.decode(outputs[0], skip_special_tokens=True).strip()

# Imprimir resultado
print("Respuesta JSON:", output_text)
```
